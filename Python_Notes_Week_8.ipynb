#W8L3 Recursion: Find 0 in a list

```python
test_case1 = [5,4,6,8,2,3,4,0,5,"mango"]
test_case2 = [1,2,3,4,5,6,7,8,9,740, "tiger"]
'''
This is a piece of code to check if a given list has 0 in it or not.
if it has "zero" in it, We return True (1), otherwise we return False (0)
'''
def check0(L):
  # if the list is empty, return False
  if (len(L)==0):
    return 0
  if (L[0]==0):
    # if the first element of the list is zero then return 1
    return 1
  else:
    return check0(L[1:len(L)])    # we are omiting the first element and feeding the rest of the list here

# the above line 15 simply outsources L[1:len(l)]

print(check0(test_case1))
print(check0(test_case2))


# Note: here 0 implies False and 1 implies True

'''
please note, this is not an efficient code, we will see this later
'''
```

#W8l4: Sorting Recursively


```python
test_case1 = [5,4,6,8,2,3,4,0,5]
test_case2 = [1,2,3,4,5,6,7,8,9,740]

def mini(L):
  '''finds the minuimum element in the list L'''
  mini = L[0]
  for x in L:
    if (x < mini):
      mini = x
  return mini

def SortList(L):
  '''recursively sort the list L'''
  if (L==[]) or (len(L) == 1):
    return L                # if the list is empty or it has only one object then return the list

  m = mini(L)
                      # m now contains the minimum most element in L
  L.remove(m)
                      # we remove that element from L
  return [m] + SortList(L)
                      # we recursively sort the smaller list



print(SortList(test_case1))
print(SortList(test_case2))
```

#W8L7 Binary Search Implementation



```python
def obvious_search(L,k):
  ''' Check if a given number k is present in a list or not. This function was authored by S.R.S. Iyengar'''
  for x in L:
    if x == k:
      return 1
  return 0
```

```python
''' A questtion: Can we write a piece of code that searches for  a given object in the list L faster than the obvious algorithm that is given above'''

def binary_search(L,k):
  '''this function is an alternative for the obvious search. It does exactly what is expected from the obvious search, but in an efficient way.
  this method is popularly known as the binary search'''

  # we want to shrink my list
  #we will do that using a while loop

  begin = 0 # first element of the list in L. L[0]
  end = len(L)-1    # the last element in L is in len(L). L[len(L)-1]

  # use a while loop to look at the list and keep halving it

  while (end - begin > 1):
    # we will handle the case when the number of elements is less than or equal to 1
    # compute the mid which is the mid point of begin to end

    mid = (begin + end)//2

    # if mid is indeed k, then we return True and stop the code.
    if (L[mid]== k):
      return 1

# if the middle element is greater than k, then cut the right side and retain the left side.
```
