```python
df2 = pd.read_excel(r"C:\Users\sarth\OneDrive\Documents\My Projects\Project2_sem7\Plots_for_literature\Sol_GOR_Plots.xlsx", sheet_name = "Goldie-1_BHS")
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(x = df2['p'],y = df2['Rs'], color = "blue", label = 'Sol GOR', legend = False)
sns.lineplot(x = df2['p'],y = df2['Rs'], color = "blue")
plt.xlabel('Reservoir Pressure, psig')
plt.ylabel('Solution Gas-Oil Ratio Rs, SCF/STB')
plt.title('Goldie-1-BHS, Solution Gas-Oil Ratio')
plt.axvline(1775, linestyle = "--", color ="red", label = "Bubble point prerssure, Pb = 1775 psia")
plt.legend()

plt.savefig('SOL GOR.png', dpi = 1200,bbox_inches='tight')
```
