```python
fig = plt.figure(figsize = (15,10))


fig.add_subplot(4,2,1)

# residual plots for linear regression
sns.histplot(x = df['P_psia'], stat = 'count',element = "step", kde = True, hue = df['Region'], palette = ['blue', 'maroon'])
plt.xlabel('Pressure, psia')
plt.ylabel('Count')
plt.title('Measured values of Pressure in psia')
#plt.show()

########################################################

fig.add_subplot(4,2,2)

# residual plots for linear regression
sns.histplot(x = df['Rs_SCF/STB'], stat = 'count',element = "step", kde = True, hue = df['Region'], palette = ['blue', 'maroon'])
plt.xlabel("Rs Solution gas-oil ratio, SCF/STB")
plt.ylabel('Count')
plt.title('Values of Solution gas-oil ratio in SCF/STB')
#plt.show()

########################################################


########################################################
fig.add_subplot(4,2,3)

# residual plots for linear regression
sns.histplot(x = df['Oil_FVF'], stat = 'count',element = "step", kde = True, hue = df['Region'], palette = ['blue', 'maroon'])
plt.xlabel('Oil Formation Volume Factor, rbbl/STB')
plt.ylabel('Count')
plt.title('Oil Formation Volume Factor, rbbl/STB')
#plt.show()


########################################################

fig.add_subplot(4,2,4)

# residual plots for linear regression
sns.histplot(x = df['Oil_dens_g/cc'], stat = 'count',element = "step", kde = True, hue = df['Region'], palette = ['blue', 'maroon'])
plt.xlabel('Oil Density, g/cc')
plt.ylabel('Count')
plt.title('Measured Oil density in g/cc')
#plt.show()


########################################################

fig.add_subplot(4,2,5)

# residual plots for linear regression
sns.histplot(x = df['Z_factor'], stat = 'count',element = "step", kde = True, hue = df['Region'], palette = ['blue', 'maroon'])
plt.xlabel('Z-Factor')
plt.ylabel('Count')
plt.title('Experimental Gas Deviation Z-Factor')
#plt.show()


########################################################

fig.add_subplot(4,2,6)

# residual plots for linear regression
sns.histplot(x = df['Temp_F'], stat = 'count',element = "step", kde = True, hue = df['Region'], palette = ['blue', 'maroon'])
plt.xlabel('Temperature, F')
plt.ylabel('Count')
plt.title('Measured Temperature in Fahrenheit')
#plt.show()


########################################################

fig.add_subplot(4,2,7)

# residual plots for linear regression
sns.histplot(x = df['gas_spgr'], stat = 'count',element = "step", kde = True, hue = df['Region'], palette = ['blue', 'maroon'])
plt.xlabel('Gas specific gravity (air = 1)')
plt.ylabel('Count')
plt.title('Gas Specific Gravity')
#plt.show()

########################################################

fig.add_subplot(4,2,8)

# residual plots for linear regression
sns.histplot(x = df['Vis_cP'], stat = 'count',element = "step", kde = True, hue = df['Region'], palette = ['blue', 'maroon'])

plt.xlabel('Oil Viscosity, cP')
plt.ylabel('Count')
plt.title('Measured Oil Viscosity in cP')
#plt.show()


#################

fig.suptitle('Univariate histogram plots showing distribution of features', fontsize  = 16)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.12, hspace=0.55)
fig.subplots_adjust(top=0.93)
plt.show()
```
